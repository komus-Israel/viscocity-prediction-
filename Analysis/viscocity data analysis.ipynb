#### The relationship between the features and viscocity

**1. Temparature**

```python
sns.scatterplot(x = 'Temperature(in celsius)', y = 'log(viscosity) in cP', data = data)
```

```python
data['Temperature(in celsius)'].unique()
```

With **increase in temperature, the viscocity is low**. This is because, the thermal energy sets the molecules on kinetic energy and this defines their ability to flow, hence their low viscocity.

The viscocity increases with low temparature. This confirms the theory that defines the relationship between temparature and viscocity.

**2. Concentrations**

```python
def Plot_Concentration(con):
    return sns.scatterplot(con, 'log(viscosity) in cP', data = data)
```

```python
Plot_Concentration('Polymer conc(wt%)')
```

Polymers with higher concentrations have higer viscocity.

However, there are high concentrations with low viscocity. What other factor could to the effect of the concentration on the viscocity ?

1. I'll check the temparature of the high concentrations with low viscocity.

```python
sns.scatterplot('Polymer conc(wt%)', 'Temperature(in celsius)', data = data, hue='log(viscosity) in cP')
```

If we examine the datapoints in the plot showing the relationship between viscocity and polymer concentration, we'll observe that most of the datapoint have a viscocity that is less than 1000.

Most of the polymers with increased temparature have a concentration of 0.20
